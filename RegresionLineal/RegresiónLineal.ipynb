```python
# Librerías a utilizar
import matplotlib.pyplot as plt
import numpy as np
import sklearn.linear_model
```

```python
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
plt.xlabel("Horas de estudio")
plt.ylabel("Calificaciones")
plt.scatter(horas_estudio, calificaciones)
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=16)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(train_data, train_labels)
```

```python
predictions = model.predict(test_data)

X_new= [[22]]
print(model.predict(X_new))
```

```python
X_new2= [[35]]
print(model.predict(X_new2))
```

```python
X_new3= [[43]]
print(model.predict(X_new3))
```

```python
plt.scatter(train_data, train_labels, color='blue', label='Datos de entrenamiento')
plt.scatter(test_data, test_labels, color='red', label='Datos de prueba')
plt.plot(test_data, predictions, color='green', linewidth=3, label='Modelo de regresión lineal')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.title('Modelo de Regresión Lineal')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(test_labels, predictions)
print("Error Cuadrático Medio:", mse)
```
